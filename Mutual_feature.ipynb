Finding accuracy before feature selection

```python
lgbm_model=lgb.LGBMClassifier()
lgbm_score=cross_val_score(lgbm_model,X,y, cv=5, scoring='accuracy')
print(lgbm_score)
print(lgbm_score.mean())
```
