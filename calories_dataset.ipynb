```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/weight_change_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Gender'].value_counts()
```

```python
df['Gender']=df['Gender'].map({'M':1, 'F':0})
```

```python
df['Physical Activity Level'].value_counts()
```

```python
df=pd.get_dummies(df, columns=['Physical Activity Level'], prefix='PAL')
df.head()
```

```python
df['Sleep Quality'].value_counts()
```

```python
df=pd.get_dummies(df, columns=['Sleep Quality'])
df.head()
```

```python
df.columns
```

```python
cols=['PAL_Lightly Active', 'PAL_Moderately Active', 'PAL_Sedentary',
       'PAL_Very Active', 'Sleep Quality_Excellent', 'Sleep Quality_Fair',
       'Sleep Quality_Good', 'Sleep Quality_Poor']
df[cols]=df[cols].astype(int)
```

```python
df.head()
```

Now we do feature engineering

```python
# importing libraries
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
# creating 2 models
rf_model=RandomForestRegressor()
elastic=ElasticNet()
```

```python
# assigning X and y
X=df.drop(columns='Final Weight (lbs)')
y=df['Final Weight (lbs)'].copy()
```

```python
rf_cross_score=cross_val_score(rf_model, X, y,cv=5,scoring='r2')
elastic_cross_score=cross_val_score(elastic, X, y,cv=5,scoring='r2')

print(f'Random forest cross validation score mean:',rf_cross_score.mean())
print(f'ElasticNet cross validation score mean:',elastic_cross_score.mean())
```

```python
df.head()
```

```python
df2=df.copy()
df2['Weight_per_age']=df2['Current Weight (lbs)']/df2['Age']
df2['Dail_cal_per_age']=df2['Daily Calories Consumed']/df2['Age']
df2.head()
```

```python
# models
rf_model=RandomForestRegressor()
elastic=ElasticNet()


# X and y
X=df2.drop(columns='Final Weight (lbs)')
y=df2['Final Weight (lbs)'].copy()


# cross validation scores
rf_cross_score=cross_val_score(rf_model, X, y,cv=5,scoring='r2')
elastic_cross_score=cross_val_score(elastic, X, y,cv=5,scoring='r2')


# final outputs
print(f'Random forest cross validation score mean:',rf_cross_score.mean())
print(f'ElasticNet cross validation score mean:',elastic_cross_score.mean())
```
