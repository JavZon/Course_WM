```python
df['person_education']=df['person_education'].replace({'High School':'High_School'})
```

```python
df.columns = [col.replace(' ', '_') for col in df.columns]
classification = setup(data=df,
                       target='loan_status',
                       session_id=123,
                       normalize=True,
                       transformation=True,
                       feature_selection=True,
                       remove_outliers=True)
```

```python
best_model=compare_models()
```

```python
final_model=finalize_model(best_model)
```

```python
y_pred=predict_model(final_model, data=df)
```
